```python
print(student.head());
print(student.shape);
print(student.columns.tolist());
```

```python
print(student.describe());
```

```python
print(student.info());
```

```python
plt.plot(student['math score'].head(10))
plt.title('Math Scores')
plt.xlabel('Student')
plt.ylabel('Score')
plt.show();
```

```python
scores = student['math score'].head(10)
x=range(1, 11)
plt.plot(x, scores, marker='o')
plt.title('Math Scores')
plt.xlabel('Student')
plt.ylabel('Score')
plt.show();
```

```python
gender_counts = student['gender'].value_counts()
plt.bar(['Female','Male'],gender_counts.values)
plt.title('Gender Count')
plt.show();
```

```python
avg_math = student.groupby('gender')['math score'].mean()
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
plt.bar(avg_math.index, avg_math.values, color=['lightblue','salmon'])
plt.title('Average Math Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show();
```

```python
plt.scatter(student['math score'], student['reading score'])
plt.grid(True)
plt.xlabel('Math')
plt.ylabel('Reading')
plt.title('Math Scores vs Reading Scores')
plt.show();
```

```python
sample = student.head(100)
plt.scatter(sample['math score'], sample['reading score'], alpha=0.5)
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show();
```

```python
plt.hist(student['math score'])
plt.title('Math Score Distribution')
plt.show();
```

```python
plt.hist(student['math score'], bins=20, color='green')
plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Math Score')
plt.show();
```

```python
test_prep=student['test preparation course'].value_counts()
plt.pie(test_prep.values, labels=['None','Completed'])
plt.title('Test Preparation')
plt.show();
```

```python
plt.pie(test_prep.values, labels=['None','Completed'], autopct='%1.1f%%')
plt.title('Test Preparation')
plt.show();
```

```python
education_counts=student['parental level of education'].value_counts()
plt.pie(education_counts, labels=education_counts.index, autopct='%1.2f%%')
plt.title('Parental Education levels')
plt.show();
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].hist(student['math score'], bins=15)
axes[0, 0].set_title('Math Score')

student['gender'].value_counts().plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Gender')

axes[1,0].scatter(student['math score'][:50], student['reading score'][:50])
axes[1,0].set_title('Math vs Reading')

axes[1,1].plot(student['writing score'][:30])
axes[1,1].set_title('Writing Score')

plt.tight_layout()
plt.show();
```

```python
sns.histplot(student['math score'])
plt.title('Math Score Distribution')
plt.show();

sns.histplot(student['math score'], kde=True)
plt.show()
```

```python
sns.boxplot(data=student,x='gender',y='math score')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Math Score by Gender')
plt.show()

sns.boxplot(data=student, x='lunch', y='math score')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Math Score by Lunch')
plt.show()
```

```python
sns.countplot(data=student, x='race/ethnicity')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Race/Ethnicity Count')
plt.show()
```

```python
sns.countplot(data=student, x='parental level of education')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Test Prep count')
plt.show()
```

```python
sns.barplot(data=student, x='parental level of education', y='math score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.kdeplot(data=student, x='math score', hue='gender')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title('Math Score by Gender')
plt.show()
```

```python
scores=student[['math score','reading score','writing score']]
correlation=scores.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot=True)
plt.title('Score Correlation')
plt.show()
```
